## Introduction to groupby()

```python
sectors = data.groupby('Sector') #  grouping by sector
# print(len('sectors')) # to see how many groups are there
sectors.size().sort_values(ascending=False) # to see how many companies are there in each sector  
```

```python
sectors.first() # to see the first company in each sector
```

```python
sectors.last() # to see the last company in each sector
```

```python
sectors.groups # to see the groups, returns a dictionary mapping group names to row indices
# data.iloc[51,:]
```

```python
sectors.get_group('Energy') # to see all the companies in the energy sector
```
