```python
def city_count(a):
    mask = df["city"].str.lower() == a.lower()
    return df[mask].shape[0]    

city_count("hyderabad")
```

Q. find number of row where city == hyderabad and dl_applied == 1, Using function .

```python
def city_count(a): # 
    mask0 = df["city"].str.lower() == a.lower()
    mask1 = df["dl_applied"] == 1
    mask = mask0 & mask1
    return df[mask].shape[0]  

city_count("hyderabad")
```

```python
## Find the count of team present in team1 and team2 column.. 
def Get_City_Name_count(team):
    mask = df['team1'].str.lower() == team.lower()
    mask1 = df['team2'].str.lower() == team.lower()
    return df[mask | mask1].shape[0]

Get_City_Name_count('sunrisers hyderabad')
```

# â›”Heandling Duplicates value using pandas

```python
# df.drop_duplicates(subset=['city']) # duplicate value remove , on the basis of city column
df.drop_duplicates(subset=['player_of_match']).shape[0] # duplicate value remove , on the basis of player_of_match column
```

```python
df.drop_duplicates(subset=['player_of_match'], keep='last').shape[0] # duplicate value remove , on the basis of player_of_match column, keep last value  
```

```python
# df.drop_duplicates(subset=['city','winner']).shape[0]
df.drop_duplicates(subset=['team1','team2']).shape[0] # duplicate value remove , on the basis of team1 and team2 column, keep first value by default 
```

Who won in each season?

```python
won = df.drop_duplicates(subset=['season'],keep="last")
won[['season','winner']].sort_values('season').plot(kind="bar")
```
