```python
import seaborn as sns
titanic = sns.load_dataset('titanic')
titanic
```

```python
titanic.shape
```

```python
titanic['survived'].value_counts()
```

```python
mask = titanic['survived'] == 1
mask1 = titanic['sex'] == 'male'
mask2 = titanic['age'] < 18
titanic[mask & mask1 & mask2].groupby('alive')['alive'].count()
```

```python
titanic.shape
```

```python
titanic.dropna().shape # dropping all the rows with null values 
```

```python
titanic.dropna(how='all').shape # dropping all the rows with all null values
```

```python
titanic.dropna(subset=['age']).shape
```

```python
titanic['deck'].unique()
```

```python
mask = titanic['deck'].isin(['C', 'E', 'G', 'D', 'A', 'B', 'F'])
mask1 = titanic['survived'] == 1
mask2 = titanic['adult_male'] == True
mask3 = titanic['alone'] == True
filtered = titanic[mask & mask1 & mask2 & mask3].dropna()
filtered['deck'].value_counts().plot(kind='bar', title='Survivors by Deck')
```

```python
mask = titanic['deck'].isin(['C', 'E', 'G', 'D', 'A', 'B', 'F'])
mask1 = titanic['survived'] == 1
filtered = titanic[mask & mask1].dropna()
filtered['deck'].value_counts().plot(kind='bar', title='Survivors by Deck')
```

```python
mask = titanic['deck'].isin(['C', 'E', 'G', 'D', 'A', 'B', 'F'])
mask1 = titanic['survived'] == 1
filtered = titanic[mask & mask1].dropna()
filtered['deck'].value_counts()
sns.barplot(x=filtered['deck'].value_counts().index, y=filtered['deck'].value_counts().values)
```

```python
sns.scatterplot(data=titanic, x='fare', y='age', hue='who') # hue is used to add another dimension to the plot
```

```python
child = titanic[titanic['who'] == 'child']
survival = child[titanic['survived'] == 1]
sns.barplot(data=survival, x='class', y='age', hue='sex')
```
