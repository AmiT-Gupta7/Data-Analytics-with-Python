```python
## Filter the data on the basis of city...
def count(city):
    mask = data['city'] == city
    return data[mask].shape[0]

count("Bangalore") # function call
```

```python
## Function to filter data on the basis of city and year greater than 2010
def year_greater_2010(city):
    mask1 = data['city'].str.lower() == city.lower()
    mask2 = data['date'] > '1010-01-01' # date is in string format
    return data[mask1 & mask2].shape[0] # shape[0] = row count 
     
year_greater_2010("hyderabad")
```

```python
## Find the count of team present in team1... 
def Get_City_Name_count(team):
    mask = data['team1'].str.lower() == team.lower()
    mask1 = data['team2'].str.lower() == team.lower()
    return data[mask | mask1].shape[0]

Get_City_Name_count('sunrisers hyderabad')
```
