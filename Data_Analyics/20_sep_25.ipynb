```python
# plotting multiple plots together...
plt.plot(data5['index'],data5['V Kohli'],color="#")
plt.plot(data5['index'],data5['RG Sharma'],color='#')
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.title("Rohit Shrama Vs Virat Kohli")
```

```python
# plotting multiple plots together...
plt.plot(data5['index'],data5['V Kohli'],linestyle='solid',linewidth=3) #linestyle: solid, dashed, dotted, dashdot, linewidth: 1,2,3
plt.plot(data5['index'],data5['RG Sharma'],linestyle='dashdot',linewidth=2)
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.title("Rohit Shrama Vs Virat Kohli")
```

```python
# plotting multiple plots together...
plt.plot(data5['index'],data5['V Kohli'],linestyle='solid',linewidth=3,marker='*',markersize=10) # marker : o,*,.,^,s, markersize: 5,10,15
plt.plot(data5['index'],data5['RG Sharma'],linestyle='dashdot',linewidth=2,marker='o',markersize=5)
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.title("Rohit Shrama Vs Virat Kohli")
```

```python
# plotting multiple plots together...
plt.plot(data5['index'],data5['V Kohli'],linestyle='solid',linewidth=3,marker='*',markersize=10,label='Virat') 
plt.plot(data5['index'],data5['RG Sharma'],linestyle='dashdot',linewidth=2,marker='o',markersize=5,label='Rohit') # label is used to show the name in legend
plt.xlabel("Season")
plt.ylabel("Runs Scored")
plt.title("Rohit Shrama Vs Virat Kohli")

plt.legend() # to show the legend
```
